```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('CKD.csv')
```

```python
dataset=pd.get_dummies(dataset, dtype=int, drop_first=True)
```

```python
independent= dataset [['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
dependent=dataset [['classification_yes']]
```

```python
X = independent
y = dependent
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.model_selection import GridSearchCV


param_grid = {
    'alpha': [0.1, 0.5, 1.0, 1.5, 2.0]  
}
         
grid = GridSearchCV(MultinomialNB(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
grid.fit(X, y) 
```

```python
from sklearn.naive_bayes import ComplementNB
```

```python
from sklearn.model_selection import GridSearchCV


param_grid = {
    'alpha': [0.1, 0.5, 1.0, 1.5, 2.0],  
      
}
         
grid = GridSearchCV(ComplementNB(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
grid.fit(X, y)
```

```python
re=grid.cv_results_
grid_predictions = grid.predict(X) 
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y, grid_predictions)
from sklearn.metrics import classification_report
clf_report = classification_report(y, grid_predictions)   
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(dependent,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```
